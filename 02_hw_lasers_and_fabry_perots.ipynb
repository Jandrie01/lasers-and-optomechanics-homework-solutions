# Geometric Series Fabry-Perot

## Part A: 
Rederive the Fabry-Perot intracavity electric field $E_\mathrm{cav}$ using the fact that the infinite geometric series
\begin{align}
\sum_{n=0}^\infty x^n = 1 + x + x^2 + \cdots x^n + \cdots = \dfrac{1}{1 - x} \qquad \text{iff} |x| < 1
\end{align}

*Hint 1:* Set up some contributing fields $E_n$ for $n$ round-trips.
## Part A  Solution: 
The electric field in the cavity of length L with N round trips is, 
$$E_{cav n}= t_1 E_{in}e^{-ikL}(r_1 r_2 e^{-2ikl})^{n-1}.$$
Using the geometric series, assuming $r_1 r _2 e^{-2ikl}$ is small,
$$E_{cav}= \frac{ t_1 E_{in}e^{-ikL}}{1- r_1 r_2 e^{-2ikl}}.$$

## Part B:
Draw a plot of the first couple of electric fields $E_n$,

## Part B Solution:

```python
#Usinf ecav_fp function from Fabry-Perot Cavity BUT modified it to have the E_cavn version from part A
def ecav_roundtrip_n(n, phis, ein, trans1, trans2, loss1=0, loss2=0):
    """Calculates the field contribution from the nth round trip.
     phis: float
        Round-trip phase phi = 2 k L [radians]
    ein: float
        Input electric field amplitude [sqrt watts]
    trans1: float
        Input mirror power transmission T1
    loss1: float
        Input mirror loss L1
    loss2: float
        End mirror loss L2      
    """
    # Set up amplitude reflectivities
    t1 = np.sqrt(trans1)
    t2 = np.sqrt(trans2)
    r1 = np.sqrt(1 - loss1 - trans1)
    r2 = np.sqrt(1 - loss2 - trans2)
    
    ecav_n = ein * t1 * np.exp(-1j * phis/2) * (r1 * r2 * np.exp(-1j * phis))**(n-1)
    return ecav_n
```

```python
#Parameters - from example in lecture
trans1 = 0.25
trans2 = 0.25
ein = 1  # sqrt(Watts)
nn = 10  #number of round trip fields

#two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

#On resonan suplot
phi_resonance = 0

#Ecav total n->infty
n_terms = 100 
ecav_total = np.sum([ecav_roundtrip_n(n, phi_resonance, ein, trans1, trans2) for n in range(1, n_terms + 1)])

#Plot individual n round trip contributions
full_field_real = 0
full_field_imag = 0

for n in range(1, nn + 1):  
    ecav_n = ecav_roundtrip_n(n, phi_resonance, ein, trans1, trans2)
    real_n = np.real(ecav_n)
    imag_n = np.imag(ecav_n)
    
    ax1.plot([full_field_real, full_field_real + real_n],[full_field_imag, full_field_imag + imag_n],'o-', label=f"n={n}", markersize=4)
    
    full_field_real += real_n
    full_field_imag += imag_n

# Plot total field
ax1.plot([0, np.real(ecav_total)], [0, np.imag(ecav_total)], 'k--', linewidth=2, color='r', label=r"$E_\mathrm{cav,total}$")
ax1.plot(np.real(ecav_total), np.imag(ecav_total), 'k*', color='r', markersize=15)
ax1.set_xlabel("Real")
ax1.set_ylabel("Imaginary")
ax1.grid(True, alpha=0.3)
ax1.set_title(r"On Resonance ($\phi = 0$)")
ax1.legend()

#Off resonance
phi_off = 0.3 

#Total field
ecav_total_off = np.sum([ecav_roundtrip_n(n, phi_off, ein, trans1, trans2) for n in range(1, n_terms + 1)])


for n in range(1, nn + 1):  # n goes from 1 to nn
    ecav_n = ecav_roundtrip_n(n, phi_off, ein, trans1, trans2)
    real_n = np.real(ecav_n)
    imag_n = np.imag(ecav_n)
    
    ax2.plot([full_field_real, full_field_real + real_n], 
             [full_field_imag, full_field_imag + imag_n], 
             'o-', label=f"n={n}", markersize=4)
    
    full_field_real += real_n
    full_field_imag += imag_n

# Plot total field
ax2.plot([0, np.real(ecav_total_off)], [0, np.imag(ecav_total_off)], 'k--',color='r', linewidth=2, label=r"$E_\mathrm{cav,total}$")
ax2.plot(np.real(ecav_total_off), np.imag(ecav_total_off), 'k*', color='r', markersize=15)

ax2.set_xlabel("Real")
ax2.set_ylabel("Imaginary")
ax2.grid(True, alpha=0.3)
ax2.set_title(rf"Off Resonance ($\phi = {phi_off:.2f}$ rad)")
ax2.legend()


plt.tight_layout()
plt.show()
```

2. How much time $t$ must elapse for the $n$th term of $E_n$ to start contributing?  Write an expression for $n$ in terms of $t$ and $\tau_{rt}$.


For the first couple of round trips,

Initial field enters cavity n=1 --- t=0


First round trip   n=2 --- t =$\tau_{rt}$

n=3 --- t =2 $\tau_{rt}$

So, for n t=(n-1)$\tau_{rt}$. Therefore, $n= \frac{t}{\tau_{rt}} +1$.

3. Using a partial geometric series, what is the buildup for the cavity $E_\mathrm{cav}(n)$ after $n$ terms are summed together?
For build up in the cavity, resonance condtion $e^{-ikL} = 1$. Then $E_\mathrm{cav}(n) = t1 E_{in}(r_1 r_2)^{n-1}$.

Since $(r_1 r_2)<<1$ with the partial geometric series,
$$E_\mathrm{cav}(n) = t1 E_{in}\frac{1-(r_1 r_2)^n}{1-r_1 r_2}.$$
4. Using the model $1 - \exp(-t / \tau_\mathrm{storage})$, calculate the cavity storage time $\tau_\mathrm{storage}$.
$E_\mathrm{cav}(n)$  in terms or t and $\tau_\mathrm{storage}$,
$$E_\mathrm{cav}(t,\tau_\mathrm{storage}) = t1 E_{in}\frac{1-(r_1 r_2)(r_1 r_2)^\frac{t}{\tau_{rt}}}{1-r_1 r_2}.$$

Assuming $r_1r_2 << 1 $ .
We need, 
\begin{align*}
1- (r_1 r_2)^{\frac{t}{\tau_{rt}}} &=1 - e^{-\frac{t}{\tau_{\mathrm{storage}}}}\\
\ln\left((r_1 r_2)^{\frac{t}{\tau_{rt}}}\right) &= \ln\left(e^{-\frac{t}{\tau_{\mathrm{storage}}}}\right)\\
\frac{t}{\tau_{rt}} \ln(r_1 r_2) &= -\frac{t}{\tau_{\mathrm{storage}}}
\end{align*}

Therefore:

$$\tau_{\mathrm{storage}} = -\frac{\tau_{rt}}{\ln(r_1 r_2)}$$

When $r_1 r_2 <<1$ \ln(r_1 r_2) \approx 1$.

5. Compare your result to the cavity pole $\nu_p$.

$$\nu_p = \frac{FSR}{F} = \frac{c/2L \cdot (1-r_1 r_2)}{\pi}=\frac{1-r_1 r_2}{\tau_{rt} \pi}$$

When $ r_1 r_2 <<1$,
$$\nu_p = \frac{FSR}{F} \approx \frac{1}{\tau_\mathrm{storage} \pi}.$$

# Finesse and Loss in a Fabry-Perot
A very convenient relationship between total loss in a cavity $\mathcal{L}_\mathrm{total}$ 
and cavity finesse $\mathcal{F}$ can be calculated to be (in the high-finesse limit $\mathcal{F} \gg 1$):
\begin{align}
\mathcal{F} = \dfrac{2 \pi}{\mathcal{L}_\mathrm{total}}
\end{align}

Derive this result starting with $\mathcal{F} = \dfrac{\mathrm{FSR}}{\nu_\mathrm{FWHM}}$.

*Hint 1:* Total loss includes transmission losses for both mirrors: $\mathcal{L}_\mathrm{total} = T_1 + T_2 + \mathcal{L}_1 + \mathcal{L}_2$.

*Hint 2:* Write $r = \sqrt{1 - T - \mathcal{L}}$.

*Hint 3:* Use the binomial approximation.

*Hint 4:* This paper from MIT may be helpful: [*Loss in long-storage-time optical cavities*](https://arxiv.org/abs/1310.1820)

# Finesse and Loss in a Fabry-Perot Solution:

$$\mathcal{F} = \dfrac{\mathrm{FSR}}{\nu_\mathrm{FWHM}}= \frac{\pi \sqrt{r_1 r_2}}{1-r_1 r_2}$$

$$r_1 r_2 = \sqrt{1 - T_1 - L_1} \sqrt{1 - T_2 - L_2}$$

Assuming  $T_i - L_i$ is small then with approximation,
$$r_1 r_2 \approx \left(1 - \frac{T_1 + L_1}{2}\right)\left(1 - \frac{T_2 + L_2}{2}\right)  \approx 1 - \frac{T_1 + L_1}{2} - \frac{T_2 + L_2}{2} + \frac{(T_1 + L_1)(T_2 + L_2)}{4}.$$
Loss and transmission should be low so can drop second order.
Then,
$$1- r_1 r_2 \approx   \frac{T_1 + L_1 T_2 + L_2}{2} \approx \frac{\mathcal{L}_\mathrm{total}}{2}.$$
Now,
$$\sqrt{r_1 r_2} \approx 1- \frac{\mathcal{L}_\mathrm{total}}{4}.$$
With high finess $\mathcal{L}_\mathrm{total} << 1$ so can aproximate $\sqrt{r_1 r_2} \approx 1 $.
Hence,
$$\mathcal{F} = \dfrac{2 \pi}{\mathcal{L}_\mathrm{total}}.$$

# Finesse and Gain in a Fabry-Perot

## Part A:
Assuming that we have a *critically-coupled* Fabry-Perot cavity,\
what is the comparison between the finesse $\mathcal{F}$ and power gain $G_\mathrm{cav}$?
## Part A Solution:
$$\frac{\mathcal{F}}{G_\mathrm{cav}} =\frac{\dfrac{2 \pi}{\mathcal{L}_\mathrm{total}}}{\frac{T_1}{(1-r_1 r_2)^2}} \approx \frac{\dfrac{2 \pi}{\mathcal{L}_\mathrm{total}}}{\frac{T_1}{\mathcal{L}_\mathrm{total}^2 /4}} = \frac{ \pi \mathcal{L}_\mathrm{total}}{2 T_1} $$
Assuming lossless mirrors we know,$T_1 =1 -R_1  = 1-r_1^2 $. Since this is a critically-coupled cavity, $r_1=r_2$. So, $T_1 =1 -R_1 = 1-r_1 r_2 \approx \frac{\mathcal{L}_\mathrm{total}}{2}.$
Hence, 
$$ \frac{\mathcal{F}}{G_\mathrm{cav}} = \pi . $$

## Part B:
Repeat Part A above for an *over-coupled* cavity such that $r_2 \approx 1$.
Does the relationship change?

For overcoupled $r_1 < r_2$, and with the condition $r_2 \approx 1$ we have $r_1< 1$.
The relationship will change.
$$\frac{\mathcal{F}}{G_\mathrm{cav}} =\frac{\dfrac{2 \pi}{\mathcal{L}_\mathrm{total}}}{\dfrac{T_1}{(1-r_1)^2}}=\frac{\dfrac{2 \pi}{\mathcal{L}_\mathrm{total}}}{\dfrac{1-r_1^2}{(1-r_1)^2}} =\frac{\dfrac{2 \pi}{\mathcal{L}_\mathrm{total}}}{\dfrac{(1-r_1)(1+r_1)}{(1-r_1)^2}} = \frac{2\pi}{\mathcal{L}_\mathrm{total}} \cdot \frac{1-r_1}{1+r_1}$$

```python

#Adding Eref definition from lecture 
def erefl_fp(phis, ein, trans1, trans2, loss1=0, loss2=0):
    """Calculates the reflected field Erefl for a Fabry-Perot cavity.
    phis: float
        Round-trip phase phi = 2 k L [radians]
    ein: float
        Input electric field amplitude [sqrt watts]
    trans1: float
        Input mirror power transmission T1
    loss1: float
        Input mirror loss L1
    loss2: float
        End mirror loss L2      
    """
    # Set up amplitude reflectivities
    t1 = np.sqrt(trans1)
    t2 = np.sqrt(trans2)
    r1 = np.sqrt(1 - loss1 - trans1)
    r2 = np.sqrt(1 - loss2 - trans2)
    
    erefl = ein * (r1 - (r1**2 + t1**2) * r2 * np.exp(-1j * phis)) / (1 - r1 * r2 * np.exp(-1j * phis))
    return erefl

# Parameters to simulate Gires-Tournois Interferometer
trans1_values = [0.1, 0.15, 0.2, 0.25, 0.3]  # Different T1 values to plot
trans2 = 0.0  # R2 = 100% 
loss1 = 0.0  
loss2 = 0.0   
ein = 1.0     


# round-trip phases in relation to frequency
# phi = 2kL = 4pi nu L/c,
phis = np.linspace(0, 4*np.pi, 1000)

# Convert round-trip phase to frequency correctly
L = 0.1 #cm
frequencies = phis * const.c / (4 * np.pi * L)


# Bode plots - magnitude, reflection phase vs frequencys and reflectionphase vs roundtrip phase
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

# Loop through different T1 values 
for tt in trans1_values:
    # Calculate reflected field
    erefl = erefl_fp(phis, ein, tt, trans2, loss1, loss2)
    
    #Calculate magnitude and phase
    magnitude = np.abs(erefl)
    phase = np.angle(erefl)
    phase_deg_unwrap = np.degrees(np.unwrap(phase))  # Fix phase discontinuities
    
    # label for legend
    r1_percent = (1 - tt) * 100
    label_r1 = f'R1={r1_percent:.0f}% (T1={tt:.0%})'
    
    # Plot magnitude vs round-trip phase
    ax1.plot(phis, magnitude, linewidth=2, label=label_r1)

    # Plot phase vs frequency
    ax2.plot(frequencies, phase_deg_unwrap, linewidth=2, label=label_r1)
    
    # Plot phase vs round-trip phase
    ax3.plot(phis, phase_deg_unwrap, linewidth=2, label=label_r1)

# Confirming Magnitude
ax1.set_xlabel(r'Round-trip phase $\phi_s$ [radians]')
ax1.set_ylabel(r'$|E_{\mathrm{refl}}/E_{\mathrm{inc}}|$')
ax1.set_title('Gires-Tournois Reflectivity Magnitude (R2=100%)')
ax1.grid(True, alpha=0.3)
ax1.set_ylim([0, 1.2])
ax1.legend()

# Phase vs frequency plot
ax2.set_xlabel('Frequency [Hz]')
ax2.set_ylabel('Phase angle [degrees]')
ax2.set_title('Reflection Phase vs Frequency')
ax2.grid(True, alpha=0.3)
ax2.legend()


# Format phase vs round-trip phase plot
ax3.set_xlabel(r'Round-trip phase $\phi$ [radians]')
ax3.set_ylabel('Phase angle [degrees]')
ax3.set_title('Reflection Phase vs Round-trip Phase')
ax3.grid(True, alpha=0.3)
ax3.legend()


plt.tight_layout()
plt.show()
```
